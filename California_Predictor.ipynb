```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("precos_casa_california.csv")
print(df.head())
```

```python
df.columns = ['lon','lat','house_age','rooms','bedrooms','population','households','income','price','proximity']
df
```

```python
dic = {'NEAR BAY':0,'<1H OCEAN':1,'INLAND':2,'NEAR OCEAN':3,'ISLAND':4}
df.proximity.unique()
```

```python
df=df.replace({"proximity":dic})
df
```

```python
df = df.replace(' ',np.nan)
df
```

```python
df = df.dropna()
df.count()
```

```python
df.dtypes
```

```python
df = df.apply(pd.to_numeric)
df.dtypes
```

```python
df.describe()
```

```python
#Observing correlation between columns
df['households'].corr(df['rooms'])
```

```python
np.set_printoptions(threshold=np.nan)
#remove  any above mean + 2SD and below mean - 2SD
#bedrooms
mean = df.bedrooms.mean()
std = df.bedrooms.std()
upper =  mean + 2*std
lower = mean - 2*std

df = df[(df['bedrooms']<=upper) & (df['bedrooms']>=lower)].reset_index(drop=True)

#house_age
mean = df.house_age.mean()
std = df.house_age.std()
upper =  mean + 2*std
lower = mean - 2*std

df = df[(df['house_age']<=upper) & (df['house_age']>=lower)].reset_index(drop=True)

#households
mean = df.households.mean()
std = df.households.std()
upper =  mean + 2*std
lower = mean - 2*std

df = df[(df['households']<=upper) & (df['households']>=lower)].reset_index(drop=True)

#income
mean = df.income.mean()
std = df.income.std()
upper =  mean + 2*std
lower = mean - 2*std

df = df[(df['income']<=upper) & (df['income']>=lower)].reset_index(drop=True)

#population
mean = df.population.mean()
std = df.population.std()
upper =  mean + 2*std
lower = mean - 2*std

df = df[(df['population']<=upper) & (df['population']>=lower)].reset_index(drop=True)

#price
mean = df.price.mean()
std = df.price.std()
upper =  mean + 2*std
lower = mean - 2*std

df = df[(df['price']<=upper) & (df['price']>=lower)].reset_index(drop=True)

#rooms
mean = df.rooms.mean()
std = df.rooms.std()
upper =  mean + 2*std
lower = mean - 2*std

df = df[(df['rooms']<=upper) & (df['rooms']>=lower)].reset_index(drop=True)

#Creating more collumns
df['people_per_household'] = df['population']/df['households']
df['bedrooms_per_household'] = df['bedrooms']/df['households']
df['rooms_per_household'] = df['rooms']/df['households']
df['per_capita'] = df['income']/df['population']

#Cleaning new collumns
#people_per_household
mean = df.people_per_household.mean()
std = df.people_per_household.std()
upper =  mean + 2*std
lower = mean - 2*std

df = df[(df['people_per_household']<=upper) & (df['people_per_household']>=lower)].reset_index(drop=True)

#bedrooms_per_household
mean = df.bedrooms_per_household.mean()
std = df.bedrooms_per_household.std()
upper =  mean + 2*std
lower = mean - 2*std

df = df[(df['bedrooms_per_household']<=upper) & (df['bedrooms_per_household']>=lower)].reset_index(drop=True)

#rooms_per_household
mean = df.rooms_per_household.mean()
std = df.rooms_per_household.std()
upper =  mean + 2*std
lower = mean - 2*std

df = df[(df['rooms_per_household']<=upper) & (df['rooms_per_household']>=lower)].reset_index(drop=True)

#per_capita
mean = df.per_capita.mean()
std = df.per_capita.std()
upper =  mean + 2*std
lower = mean - 2*std

df = df[(df['per_capita']<=upper) & (df['per_capita']>=lower)].reset_index(drop=True)
```

```python
%matplotlib inline
df.hist(bins=30, figsize=(15,10))
plt.show()
```

```python
from sklearn.ensemble import BaggingRegressor
from sklearn.model_selection import train_test_split
df.describe()
```

```python
y = df['price']
x = df.drop(columns=['price'])
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.4,random_state = 3)
x_test, x_valid, y_test, y_valid = train_test_split(x_test,y_test, test_size=0.8,random_state = 7)

model = BaggingRegressor(n_estimators=50)
model.fit(x_train,y_train)


score = model.score(x_train,y_train)
print(score)

score = model.score(x_test,y_test)
print(score)

score = model.score(x_valid,y_valid)
print(score)
```
